```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
from sympy import symbols, limit, sympify

x = symbols('x')
expression = (x**2 - 3*x) / (sy.sqrt(9 + x) - 3)
value = 0

symbolic_expression = sympify(expression)
x_values = np.linspace(value - 2, value + 2, 400)
y_values = [symbolic_expression.subs(x, val) for val in x_values]


plt.plot(x_values, y_values, label=r'$\frac{x^2 - 3x}{\sqrt{9 + x} - 3}$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da função')

plt.axvline(x=value, color='r', linestyle='--', label=f'x={value}')

# Calculando o limite
limit_value = sy.limit(symbolic_expression, x, value)

print("Limite:", limit_value)
plt.scatter(value, limit_value, color='purple', label=f'Limite = {limit_value}')

plt.text(value, limit_value, f'({value}, {limit_value})', fontsize=10, color='black', va='bottom')
plt.ylim(-30, -5)
plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
from sympy import symbols, limit

x = symbols('x')
expression = (x**3 - 3*x**2 + 3*x - 1)/(x**2 - 1)
value = 1

numpy_expression = sy.lambdify(x, expression, 'numpy')
x_vals = np.linspace(0.5, 1.5, 400)
x_vals = x_vals[x_vals != 1]

y_vals = numpy_expression(x_vals)

plt.plot(x_vals, y_vals, label=r'$\frac{x^3 - 3x^2 + 3x - 1}{x^2 - 1}$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da função')

plt.axvline(x=value, color='r', linestyle='--', label=f'x={value}')

# Calculando o limite
limit_value = limit(expression, x, value)

print("Limite:", limit_value)
plt.scatter(value, limit_value, color='purple', label=f'Limite = {limit_value}')

plt.text(value, limit_value, f'({value}, {limit_value})', fontsize=10, color='black', va='bottom')

plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
from sympy import symbols, limit, sympify

x = symbols('x')
expression = (x**2 - 4*x + 3)/(x**2 + 5*x + 6)
value = -3

symbolic_expression = sympify(expression)
x_values = np.linspace(value - 2, value + 2, 400)
y_values = [symbolic_expression.subs(x, val) for val in x_values]


plt.plot(x_values, y_values, label=r'$\frac{x^2 - 4x + 3}{x^2 + 5x + 6}$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da função')

plt.axvline(x=value, color='r', linestyle='--', label=f'x={value}')

# Calculando o limite
limit_value = sy.limit(symbolic_expression, x, value)

print("Limite:", limit_value)
plt.scatter(value, limit_value, color='purple', label=f'Limite = {limit_value}')

plt.ylim(-50, 50)
plt.legend()
plt.grid(True)

plt.show()
```

```python
from sympy import symbols, sympify
import numpy as np
from IPython.display import display, Math

x = symbols('x')
variable = 'x'
expression = (x**2 - 4*x + 3)/(x**2 + 5*x + 6)

func = sympify(expression)
point = float(-3)

def aproximations(limit_val, range=[0.1, 0.01, 0.001, 0.0001, 0.00001]):
    left_shift = [limit_val - d for d in range]
    right_shift = [limit_val + d for d in range]
    return left_shift, right_shift

left_shift, right_shift = aproximations(point)

left_values = [func.subs(x, ponto).evalf() for ponto in left_shift]
right_values = [func.subs(x, ponto).evalf() for ponto in right_shift]
table = "\\begin{array}{|c|c|c|c|}\n"
table += "\\hline\n"
table += "\\text{Aprox. à Esquerda} & \\text{Valor} & \\text{Aprox. à Direita} & \\text{Valor} \\\\\n"
table += "\\hline\n"
for i in range(len(left_shift)):
    table += f"{left_shift[i]:.5f} & {left_values[i]:.5f} & {right_shift[i]:.5f} & {right_values[i]:.5f} \\\\\n"
table += "\\hline\n"
table += "\\end{array}"

display(Math(table))
```
